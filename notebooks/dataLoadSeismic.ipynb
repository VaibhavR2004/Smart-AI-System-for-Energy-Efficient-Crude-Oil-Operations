```python
import segyio
import numpy as np

filename = "E:\Research_Paper\oil_gas_ai\data\processed\Seismic.sgy"

with segyio.open(filename, "r") as f:
    cube = segyio.tools.cube(f)

print("Cube shape:", cube.shape)
```

```python
import os
import matplotlib.pyplot as plt

# Path where images will be stored
save_path = r"E:\Research_Paper\oil_gas_ai\images"

# Create folder if it does not exist
os.makedirs(save_path, exist_ok=True)

# Loop through inline slices
for i in range(0, cube.shape[0], 5):

    img = cube[i, :, :]

    # Create full file path
    filename = os.path.join(save_path, f"inline_{i}.png")

    # Save image
    plt.imsave(filename, img, cmap="gray")

print("All images saved successfully!")
```
